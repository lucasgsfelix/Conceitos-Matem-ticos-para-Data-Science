# Exemplos computacionais 1: Vetores e Matrizes

## Representação computacional de vetores:

```python
vetor_a = np.array([5, 6, 7, 8])

vetor_b = np.array([5, 6, 7, 8]) * 2

print(vetor_b)

# soma de vetores
vetor_c = vetor_a + vetor_b
print(vetor_c)

# multiplicando vetor por escalar
print(vetor_c * 10)

# calculando a norma do vetor
print("Norma do vetor: ", np.linalg.norm(vetor_c))
```

## Representação computacional de matrizes:

```python
# Operações com random

## Criação de um array aleatório
array = np.random.randint(low=0, high=10, size=100)


## Criação de uma matriz aleatória
matrix_array = array.reshape(10, 10)

print("Matriz A:")
print(matrix_array)


## Multiplicação de matriz por escalar
print("Multiplicação por escalar")
print(matrix_array * 2)

## Multiplicação de matriz por matriz
matrix_b = np.random.randint(low=0, high=10, size=100).reshape(10, 10)

print("Matriz B: ")
print(matrix_b)


print("Resultante: ")
print(np.matmul(matrix_array, matrix_b))


print("Transposta da Matriz: ")
print(matrix_array.T)


# criaçao da matriz identidade
identity_matrix = np.zeros((10, 10))

print(identity_matrix)

np.fill_diagonal(identity_matrix, 1)

print("Matriz identidade de tamanho n=10")
print(identity_matrix)


print("Valor da determinante: ")
print(np.linalg.det(matrix_array))


print("Calculo de auto vetores e auto valores: ")
print(np.linalg.eig(matrix_array))
```

## Derivada Parcial

## Derivada Vetorial ou Gradiente

No cálculo vetorial o gradiente (ou vetor gradiente) é um vetor que indica o sentido e a direção na qual, por deslocamento a partir do ponto especificado, obtém-se o maior incremento possível no valor de uma grandeza a partir da qual se define um campo escalar para o espaço em consideração.


## Exemplificando dentro de ciência de dados: Algoritmos de gradiente descendente

Gradiente descendente é um algoritmo de otimização que é utilização para achar o mínimo local em uma função diferenciável (que possui uma derivada). A ideia do algoritmo é caminhar em direção oposta ao gradiente da função no ponto atual, dado que é a difreção de descida mais ingríme. Se o objetivo for encontrar o ponto que máxima a função, anda-se em direção ao gradiente.

Em geral a função $f(x)$ é uma função de erro que deve ser minimizada como o MSE (Mean Squared Error) ou o RMSE (Root Mean Squared Error).


## Exemplificando dentro de ciência de dados: SVD


<img src="Imagens/RS_exemplo1.jpg">
<center>Voltando ao sistema de recomendação</center>

- Algoritmos baseados em k-Nearest Neighbors funcionam muito bem quando temos uma base de dados pequena. Esses algoritmos capturam as interações explícitas entre items e usuários.
- E quando temos uma quantidade de dados massiva? Como proceder?
- **São utilizados algoritmos baseados em fatores latentes como SVD**
- Esse algoritmo além de capturar as interações explícitas, capturam as interações implicítas
    - Além de considerarem as interações, conseguem capturar aspectos como o tipo, gênero e coisas do tipo

<img src="Imagens/svd.png">
<center>Princípio básico de funcionamento do SVD. O SVD é um algoritmo que foca em reconstruir a matriz original identificando os items que não há um rating (fazendo a predição). Imagem retirada dos slides do professor Rodrygo Santos da UFMG.</center>

- Como o princípio do SVD se baseia na reconstrução da matriz de maneira que ela fique menor, porém o mais próximo do original, a tática utilizada é de minimização do erro residual. Considerando que a função de erro utilizada é dada pela funçaõ apresentada abaixo:

<center>$e = (r_{ui} - \hat{r}_{ui})^2$</center>

Onde $r_{ui}$ é o rating real e $\hat{r}_{ui}$ é o rating predito. Vale ressaltar que a predição é dada por $\hat{r}_{ui} = p_{u}q_{i}^T = \sum_{f=1}^k p_{uf} q_{if}$


Podemos reescrever a fórmula de erro da seguinte forma:



<center>$e_{ui} = (r_{ui} - \hat{r}_{ui})^2 = (r_{ui} - \sum_{f=1}^k p_{uf} q_{if})$</center>




<img src="Imagens/derivadas_parciais_1.png">
<center>Derivada do erro em função do termo p e do termo q</center>


<img src="Imagens/derivadas_parciais_2.png">
<center>Regras de atualização com learning rate $\alpha$</center>


ps.: O learning rate basicamente define o quão rápida o algoritmo de machine learning aprende. Neste caso o algoritmo define o quão grande serão os passos do algoritmo até chegar no ponto que minimiza o erro.

```python
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import csr_matrix

np.random.seed(0)
X_dense = np.random.rand(100, 100)
X_dense[:, 2 * np.arange(50)] = 0
X = csr_matrix(X_dense)


svd = TruncatedSVD(n_components=5, n_iter=7, random_state=42)
svd.fit(X)

print(svd.explained_variance_ratio_)

print(svd.explained_variance_ratio_.sum())

print(svd.singular_values_)
```

### Técnicas de amostragem

<img src="Imagens/tecnicas_amostragem.png">
<center>Técnicas de amostragem estatística</center>

- Amostragem aleatória simples (simple random sample): Todos os items de probabilidade semelhante de serem escolhidos
- Amostragem sistemática (systematic sample): É utilizada quando a população está naturalmente ordenada (listas telefônicas). Pega-se valores de x em x, sendo x o intervalo da amostragem, x=$\frac{N}{n}$ , sendo calculado $N$ o tamanho da população e $n$ o tamanho da mostra.
- Amostragem estratificada (stratified sample): Neste caso a população é dividida em subgrupos denominados estratos (variáveis categóricas que permitem dividir a população: sexo, renda, bairro). As amostras dentro de cada estrato devem ser mais semelhantes entre do que entre as variáveis fora do estrato. Neste caso a população amostral deve ser proporcional a população original. Exemplo: Se uma amostra tem tamanho 20% na população original, então 20% de suas instâncias serão selecionadas para a populaçao amostral.
- Amostragem por conglomerado (cluster sample): Divide a população em subgrupos e utiliza todos os membros de alguns subgrupos.


É possível confundir a análise por conglomerado com a a amostragem estratificada, para diferenciar pense que a análise estratificada utiliza alguns items de todos os subgrupos, enquanto a análise por conglomerado utiliza todos os itens de alguns subgrupos.

### Regressão linear simples


A regressão linear simples tem como objetivo ajustar uma reta entre duas variáveis quantitativas.

A reta ajustada de duas variáveis quantitativas X e Y é dado por:

# $\hat{Y} = a + BX$

Onde **a** é o intercepto e **b** é a inclinação da reta.

- Interpretação do **b**: Para cada aumento de uma unidade em **X**, temos um aumento (se **b positivo**) ou um decréscimo (se **b** negativo) médio de **b** unidades em **Y**.


Neste caso podemos calcular a e b utilizando uma métrica de erros, que minimiza o desvio do erro (exatamente como fizemos no SVD):


# $f(a, b) = \sum_{i=1}^{n} e_{i}^2 = \sum_{i=1}^n (Y_{i} - (a + bX_{i}))^2$

O objetivo agora é encontrar o valor mínimo de uma função de duas variáveis, a e b. Derivando em termos de cada uma das variáveis e igualando a zero observamos que as soluções a e b devem satisfazer as seguintes equações:

## Probabilidade

- **Experimento aleatório**: Qualquer fenômeno que gere resultado incerto. Exemplo: Jogar uma moeda duas vezes e avaliar o resultado
- **Espaço amostral ($\Omega$)**: É o conjunto de todos os resultados possíveis de um experimento aleatório. Exemplo: Ao se jogar uma moeda, pode-se obter cara ou coroa.
- **Evento**: É qualquer subconjunto do espaço amostral. 


Ilustrando isso: Considere o experimento de jogar um dado e observar o resultado. Alguns possíveis eventos desse experimento são: **A** = {Ocorrer a face 5} = {5} ou **B** = {Ocorrer face par} = {2, 4, 6}.

- O evento intereseção de A e B, denotado por $A\cap B$, é um evento em que A e B ocorrem simultaneamente
- O evento união de A e B, denotado por $A \cup B$, é um evento em que A ocorre ou B ocorre (ou ambos)
- O evento complementar A, denotado por A^c, é o evento em que A não ocorre
- O evento disjunto (ou mutuamente exclusivo) se ele não pode ocorrer simultanemanete entre A e B. Exemplo: Um mesmo dado pode ser lançado e ter os valores 4 e 5?


## Distribuições de probabilidade


### Definição de Probabilidade: Probabilidade á uma medida que quantifica a incerteza de um acontecimento futuro.

Maneiras de se medir a incerteza de um evento:

- **Método Clássico**: Essa definição se aplica quando os pontos do espaço amostral são equiprováveis. Por exemplo, um lance de dados ou cara ou coroa. Considerando que a A seja o evento de interesse:

# $P(A) = \frac{Número\ de\ casos\ favoráveis\ ao\ evento\ A}{Número\ de\ casos\ possíveis}$

- **Método Frequentista**: Este é o caso da maioria das aplicações de probabilidades de jogos de azar. Esse método considera a probabilidade de um evento ocorrer várias vezes em uma série suficientemente grande de realizações de um experimento.

# $P(A) = \frac{Número\ de\ vezes\ em\ que\ A\ ocorreu}{Número\ total\ de\ repetições\ do\ experimento}$

- **Método Subjetivo**: Cada indivíduo, baseado em informações anteriores e em sua opinião a respeito de um evento em questão, pode ter uma resposta para a probabilidade desse evento. Exemplo: Probabilidade de um paciente ter certa doença.


- **Método moderno ou axiomático**: A probabilidade de uma função P(.), que associa a cada evento do espaço amostral $\Omega$ um número real, pertence ao intervalo $[0, 1]$, satisfazendo os seguintes axiomas:
    - **Axioma 1**: $0 \leq P(A) \leq 1$
    - **Axioma 2**: $P(\Omega) = 1$
    - **Axioma 3**: Se A e B são eventos mutuamente exclusivos, isto é, $A \cap B = \emptyset$, então:
    # $P(A \cup B) = P(A) + P(B)$


### Probabilidade condicional

Probabilidade condicional surge quando deseja-se calcular a probabilide de um evento A ocorrer, sabendo que um evento B já ocorreu.

Sejam A e B dois eventos associados a um espaço amostral $\Omega$ denota-se por $P(A|B)$ a probbilidade condionada do evento A, quando o evento B tiver ocorrido. 

Dado dois eventos A e B e um mesmo espaço amostral $\Omega$, a probabilidade condicional de A, dado que ocorreu B, é representada por P(A|B) e definda por:

# $P(A|B) = \frac{P(A \cap B)}{P(B)}, P(B) > 0$

Sendo

# $P(A \cap B) = P(B) \times P(A|B)$

## Teste de hipótese


O teste de hipótese é tão somente uma regra de decisão para ACEITAR ou REJEITAR uma hipótese qualquer (uma suposição, uma afirmação), com base nos elementos amostrais.

Por exemplo, suponha que a FIAT afirme que o consumo de combustível do Pálio Fire é, em média, de 18 km/L. Uma revista decide testar essa afirmação e analisa 50 veículos obtendo uma média de 17 km/L, que é diferente da informada pelo fabricante. 

O resultado de 17km/L não garante que a afirmação do fabricante seja falsa, pois você está se baseando em dados amostrais. Para haver esta garantia só realizando um censo (toda a população), o que é teoricamente impossível. O que devemos avaliar, com auxílio do Teste de Hipótese, é se a afirmação é verdadeira ou não, com base nos dados amostrais.

Testes de hipóteses são procedimentos estatísticos que permitem tomar uma decisão entre duas ou mais hipóteses utilizando dados observados de um determinado experimento.

Hipóteses formadas:

- $H_0$ - Hipótese nula
- $H_a$ - Hipótese alternativa

Testar hipóteses consiste em decidir se aceita ou rejeita uma hipótese nula ($H_0$). Quando se rejeita a hipótese nula, automaticamente se aceita a hipótese alternativa.

### Exemplo 1: Calcular a proporção de peças defeituosas produzadas em uma máquina


Considerando que a probabilidade de a máquina produzir uma peça defeituosa é de p = 0,05

As hipóteses possíveis são:

- **Teste Bilateral**: $H_{0}: p = 0.05$, $H_a: p \neq 0.05 $
- **Teste unilateral à direita**: $H_{0}: p = 0.05$, $H_a: p > 0.05 $
- **Teste unilateral à esquerda**: $H_{0}: p = 0.05$, $H_a: p < 0.05 $

<center>Teste de hipótese em casos bilaterais</center>


### Erros em testes de hipótese:

- $\alpha$: Rejeita-se a hipótese H_0 quando na realidade ela é verdadeira ($\alpha$ é chamado de nível de significância do teste)
- $\beta$: Aceita a hipótese H_0 quando na realidade ela é falsa

Os valores de $\alpha$ e $\beta$ são probabilidades de cada um dos erros tipo I e tipo ocorrerem:

# $\alpha = P(erro\ tipo \ I) = P(rejeitar\ H_{0}\ dado \ que \ H_{0}\ é\ verdadeira)$
# $\beta = P(erro\ tipo \ II) = P(aceitar\ H_{0}\ dado \ que \ H_{0}\ é\ falso)$


### Nível de significância de um teste de hipótese

É a probabilidade máxima que aceitmos cometer um erro do tipo I ($\alpha$).

<br>
<br>

Usalmente são utilizados os valores 0.1 (10%), 0.05 (5%) e (0.01) (1%). Em geral, quanto menor a probabilidade de se cometer um erro do tipo I, maior a probabilidade de se cometer um erro do tipo II. A única forma de se reduzir as probabilidades relativas dos dois erros é aumentando o tamanho da amostra (n), pois quanto maior a amostra mais confiança pode se ter nos dados.


### Exemplo 2: 

Numa amostra de 100 peças produzidas por uma máquina foram encontradas 3 defeituosas. Utilizando um nível de significância de 5%, teste se a proporção populacional de peças defeituosas desta máquina é menor que 0.08.

**Solução:**

Hipóteses: $H_0$: p = 0.08 e $H_a:$ p < 0.08 (curva unilateral à esquerda)

<img src="Imagens/esquerda.png">
<center>Curva unilateral à esquerda. Vamos considerar o $\alpha=0.05$</center>

**Lembrando que o resultado do que fizermos tem de estar dentro da região de aceitação**

Para $\alpha=5\%$, temos que z = -1,64, o valor de z observado na amostra é:

Fórmula:

# $z_{obs} = \frac{(\hat{p} - p)}{\sqrt{\frac{p (1-p)}{n}}}$

# $z_{obs} = \frac{(0.03 - 0.08)}{\sqrt{\frac{0.08 (1-0.08)}{100}}} = -1.84$

Conclusão do teste: Como $Z_{obs} (-1.84) < z (1.64)$, rejeita-se $H_{0}$, ou seja, há indicios de que a proporção de peças defeitusas seja menor que 0.08.

### Intervalo de confiança


O intervalo de confiança é importante para indicar a margem de incerteza (ou imprecisão) frente a um cálculo efetuado. Esse cálculo usa a amostra do estudo para estimar o tamanho real do resultado na população de origem.

O cálculo de um intervalo de confiança é uma estratégia que considera a amostragem de erro. A dimensão do resultado do seu estudo e seu intervalo de confiança caracterizam os valores presumíveis para a população original.

Quanto mais estreito for o intervalo de confiança, maior é a probabilidade da porcentagem da população de estudo representar o número real da população de origem dando maior certeza quanto ao resultado do objeto de estudo.

Fórmula para o intervalo de confiança:

# $\hat{X} \pm Z \frac{s}{\sqrt{n}}$

Sendo:

- $\hat{X}$: Média amostral 
- $Z$: Valor associado com a probabilidade desejada 
- $\hat{X}$: É o desvio padrão da média amostral


#### Exemplo 1:

Encontre o número de amostras $n$, calcule a média $X$ e o desvio padrão $s$ dessas amostras. Considerando que esses valores sejam 40, 1.75m e 0.20m respectivamente.


# $1.75 \pm 1.96 \times \frac{0.20}{\sqrt{40}}$

# $1.75 \pm 0.0620$

Em conclusão, com intervalo de confiança de 95% a média populacional está entre 1.68 m e 1.81 m.
